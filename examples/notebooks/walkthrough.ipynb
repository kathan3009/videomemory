# VideoMemory walkthrough

This notebook ingests a short YouTube clip end-to-end and demonstrates every retrieval mode.

Prereqs: `uv sync --extra ml`, ffmpeg installed.

```python
import asyncio
from pathlib import Path

from videomemory.pipeline.runner import run_ingest

DATA = Path('./nb_data')
job = asyncio.run(run_ingest('https://youtu.be/BM70fDqUo3c', data_dir=DATA))
print(job.video_id)
```

```python
from videomemory.query.engine import answer_question

res = asyncio.run(answer_question(job.video_id, 'summarize the main topic', DATA))
print(res.answer)
for c in res.chunks:
    print(f'[{c.start:.1f}-{c.end:.1f}] {c.summary[:100]}')
```

```python
from videomemory.retrieval.frame_recall import recall_frames

frames = asyncio.run(recall_frames(job.video_id, query='speaker', limit=5, data_dir=DATA))
for f in frames:
    print(f.timestamp, f.score, f.why)
```
